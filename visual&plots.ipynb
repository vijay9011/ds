```python
import pandas as pd
 import matplotlib.pyplot as plt
 iris = pd.read_csv('iris.csv')
 iris

iris_1 = iris.drop(columns=['Id'])
 iris_1

random_sample = iris_1.sample(n=5, random_state=1)
 print(random_sample)

iris_1.describe()

species_summary = iris_1.groupby('Species').describe()
 print(species_summary)

iris_1.boxplot(column='PetalWidthCm', by='Species',showmeans=True,␣
 ↪meanline=True,ax=plt.gca())
 plt.title('Box Plot of Petal Width by Species')
 plt.suptitle('')
 plt.show()

plt.figure(figsize=(6, 4))
 plt.scatter(iris_1[iris_1['Species'] == 'Iris-setosa']['SepalLengthCm'],
 iris_1[iris_1['Species'] == 'Iris-setosa']['SepalWidthCm'],
 label='Iris-setosa')
 plt.scatter(iris_1[iris_1['Species'] == 'Iris-versicolor']['SepalLengthCm'],
 iris_1[iris_1['Species'] == 'Iris-versicolor']['SepalWidthCm'],
 label='Iris-versicolor')
 plt.scatter(iris_1[iris_1['Species'] == 'Iris-virginica']['SepalLengthCm'],
 iris_1[iris_1['Species'] == 'Iris-virginica']['SepalWidthCm'],
 label='Iris-virginica')
 plt.title('Sepal Length vs Sepal Width')
 plt.xlabel('Sepal Length (cm)')
 plt.ylabel('Sepal Width (cm)')
 plt.legend()
plt.show()


plt.figure(figsize=(7, 3))
 for specie in iris_1['Species'].unique():
 subset = iris_1[iris_1['Species'] == specie]
 plt.scatter(subset['PetalLengthCm'], subset['PetalWidthCm'], label=specie)
 plt.title('Petal Length vs Petal Width')
 plt.xlabel('Petal Length (cm)')
 plt.ylabel('Petal width (cm)')
 plt.legend(title='Species')
 plt.show()

species_counts = iris_1['Species'].value_counts()
 print(type(species_counts))
 plt.figure(figsize=(8, 4))
 plt.bar(species_counts.index, species_counts.values, color='skyblue',␣
 ↪edgecolor='black')
 plt.title('Count of iris_1 Species')
 plt.xlabel('Species')
 plt.ylabel('Count')
 plt.xticks(rotation=45)
 plt.show()

import seaborn as sns
 import matplotlib.pyplot as plt
 sns.countplot(data=iris_1, x='Species')
 plt.title('Count of iris_1 Species')
 plt.xlabel('Species')
 plt.ylabel('Count')
 plt.legend()
 plt.show()

bins=[1,2,3,4,5,6,7,8]
 n, bn, patches = plt.hist(iris_1['PetalLengthCm'], bins=bins, color='skyblue',␣
 ↪edgecolor='black')
 for patch in patches:
 height = patch.get_height()
 plt.text(patch.get_x() + patch.get_width() / 2, height, f'{int(height)}',
 ha='center', va='bottom')
 plt.title('Distribution of Petal Length')
 plt.xlabel('Petal Length (cm)')
 plt.ylabel('Frequency')
 plt.show()

import seaborn as sns
 import matplotlib.pyplot as plt
 plt.figure(figsize=(8, 6))
 sns.histplot(iris_1['SepalWidthCm'], bins=10, kde=True, color='skyblue',␣
 ↪edgecolor='black')
 plt.title('Distribution of Petal Width in Iris Dataset')
 plt.xlabel('Petal Width (cm)')
 plt.ylabel('Frequency')
 plt.show()

iris_1_numeric = iris_1.drop(columns=['Species'])
 correlation_matrix = iris_1_numeric.corr()
 plt.figure(figsize=(8, 6))
 sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
 plt.title('Correlation Heatmap of iris_1 Dataset')
 plt.show()

iris_1_numeric = iris_1.drop(columns=['Species'])
 sns.pairplot(iris_1, hue='Species')
 plt.show()



```
